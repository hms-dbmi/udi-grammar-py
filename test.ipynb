```python
# this is needed to import the local udi-grammar-py functions
import sys
sys.path.insert(0, "./src")

# import dependencies
from udi_grammar_py import Chart, Op, UDIWidget
```

```python
# Create spec using the data URI
chart = (
    Chart()
    .source("donors", "./data/example_donors.csv")
    .mark("point")
    .x(field="weight", type="quantitative")
    .y(field="height", type="quantitative")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```

```python
# Create spec using the data URI
chart = (
    Chart()
    .source("donors", "./data/example_donors.csv")
    .rollup({"count": Op.count()})
    .mark("bar")
    .x(field="count", type="quantitative")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```

```python
# Create spec using the data URI
chart = (Chart()
    .source("samples", "./data/example_samples.csv")
    .groupby(["organ", "organ_condition"])
    .rollup({"count": Op.count()})
    .mark("bar")
    .x(field="count", type="quantitative")
    .y(field="organ", type="nominal")
    .color(field="organ_condition", type="nominal")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```

```python
# Create spec using the data URI
chart = (Chart()
    .source("donors", "./data/example_donors.csv")
    .mark("point")
    .x(field="weight", type="quantitative")
    .y(field="height", type="quantitative")
    .size(field="age", type="quantitative")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```

```python
# ARC marking
chart = (Chart()
    .source("samples", "./data/example_samples.csv")
    .groupby(["organ"])
    .rollup({"count": Op.count()})
    .mark("arc")
    .color(field="organ", type="nominal")
    .theta(field="frequency", type="quantitative")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```

```python
# ROW marking
chart = (
    Chart()
    .source("donors", "./data/example_donors.csv")
    .mark("row")
    .text(field="*", type="nominal", mark="text")
    .to_json(pretty=True)
)

# Render widget
UDIWidget(spec_json=chart)
```
